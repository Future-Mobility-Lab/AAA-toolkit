```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

import pandas as pd
from sklearn.preprocessing import LabelEncoder

def preprocess_data(df):
    """
    Preprocesses the input dataframe by encoding non-numeric columns using label encoding, 
    encoding binary/boolean values using label encoding, and replacing NaN values 
    with the mean value of the column for numerical columns and with the mode value for non-numerical columns.

    Args:
        df (pandas.DataFrame): The input dataframe to be preprocessed.

    Returns:
        pandas.DataFrame: The preprocessed dataframe.
    """
    # Define a function to encode non-numeric columns using label encoding
    def encode_non_numeric_columns(df):
        for column in df.columns:
            if df[column].dtype == object:
                encoder = LabelEncoder()
                df[column] = encoder.fit_transform(df[column].astype(str))
            elif df[column].dtype == bool:
                encoder = LabelEncoder()
                df[column] = encoder.fit_transform(df[column])
        return df
    
    # Encode non-numeric columns using the function
    df_encoded = encode_non_numeric_columns(df)
    
    # Replace NaN values with the mean value of the column for numerical columns
    # and with the mode value for non-numerical columns
    for column in df_encoded.columns:
        if df_encoded[column].dtype == 'float64' or df_encoded[column].dtype == 'int64':
            df_encoded[column] = df_encoded[column].fillna(df_encoded[column].mean())
        else:
            df_encoded[column] = df_encoded[column].fillna(df_encoded[column].mode()[0])

    return df_encoded

```

```python
import pandas as pd
from sklearn.model_selection import KFold, cross_validate
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.svm import SVR
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import make_scorer, mean_absolute_error, mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
from tabulate import tabulate
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    eps = 1e-10
    return np.mean(np.abs((y_true - y_pred) / (y_true + eps))) * 100

def create_nn_model():
    model = Sequential()
    model.add(Dense(128, activation='relu', input_dim=X.shape[1]))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
    return model

X = XP
Y = Y_C.astype(float)

models = {
    'KNN': KNeighborsRegressor(),
    'Random_Forest': RandomForestRegressor(),
    'XGBoost': XGBRegressor(),
    'Linear_Regression': LinearRegression(),
    'LightGBM': LGBMRegressor(),
    'SVM': SVR(),
    'GBDT': GradientBoostingRegressor(),
    'CatBoost': CatBoostRegressor(verbose=0),
    'Neural_Network': KerasRegressor(build_fn=create_nn_model, epochs=100, batch_size=32, verbose=0),
    'Ridge_Regression': Ridge()
}

metrics = {
    'MAE': make_scorer(mean_absolute_error),
    'RMSE': make_scorer(mean_squared_error, squared=False),
    'MAPE': make_scorer(mean_absolute_percentage_error, greater_is_better=False)
}

n_folds = 5

TARGETS={'Reported': Y_A, 'Manual': Y_B, 'Estimated': Y_C}

results = {}

for target_name, target_values in TARGETS.items():
    Y = target_values.astype(float)
    for model_name, model in models.items():
        cv_results = cross_validate(model, X, Y, cv=KFold(n_splits=n_folds, shuffle=True, random_state=42), scoring=metrics)
        for metric_name in metrics:
            key = f'{metric_name} ({model_name}) ({target_name})'
            results[key] = np.round(cv_results[f'test_{metric_name}'].mean(), 2)
            if metric_name == 'MAPE':
                results[key] = -np.round(cv_results[f'test_{metric_name}'].mean(), 2)

print(results)
```

```python
import pandas as pd

# Organize results into a nested dictionary structure
organized_results = {}
for key, value in results.items():
    metric, model, target = key.split(" ", 2)
    if target not in organized_results:
        organized_results[target] = {}
    if model not in organized_results[target]:
        organized_results[target][model] = {}
    organized_results[target][model][metric] = value

# Create a list of dictionaries, one for each target
table_data = []
for target, model_results in organized_results.items():
    row = {"Target": target}
    for model, metric_results in model_results.items():
        for metric, value in metric_results.items():
            row[f"{metric} ({model})"] = value
    table_data.append(row)

# Convert the list of dictionaries to a pandas DataFrame
results_df = pd.DataFrame(table_data)

print(results_df.T)
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(12, 6))
for i, col in enumerate(['gran', 'kernel_size', 'quantile']):
    axs[i].boxplot(df[col], vert=False)
    axs[i].set_title(col.capitalize())
plt.tight_layout()
plt.show()

# create boxplots for columns 'selectivity', 'metric', and 'f1_score'
fig, axs = plt.subplots(ncols=3, figsize=(12, 6))
for i, col in enumerate(['selectivity', 'metric', 'f1_score']):
    axs[i].boxplot(df[col], vert=False)
    axs[i].set_title(col.capitalize())
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4.5,3))
sns.boxplot(x="metric", y="f1_score", data=df[df.metric!=7][df.metric!=8])
plt.grid()
plt.xlabel("Metric")
plt.ylabel("F1 Score")

plt.title("F1 Score by Metric")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# create the dictionary of distance metrics
metric_dict = {
    0: "Bray-Curtis",
    1: "Canberra",
    2: "Chebyshev",
    3: "Manhattan",
    4: "Correlation",
    5: "Cosine",
    6: "Euclidean"
}

# create the plot
plt.figure(figsize=(3.5,3.5))
sns.boxplot(x="metric", y="f1_score", data=df[df.metric!=7][df.metric!=8])
plt.grid()
plt.xlabel("Metric",fontsize=12)
plt.ylabel("F1 Score",fontsize=12)
# plt.title("F1 Score by Metric")

# set the x tick labels and rotate them
plt.xticks(range(len(metric_dict)), metric_dict.values(), rotation=90)
plt.tight_layout()
# save the plot as a PDF file
plt.savefig("f1_score_by_metric.pdf", format="pdf")

plt.show()
```

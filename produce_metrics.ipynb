```python
def iou(start1, end1, start2, end2):
    # determine the intersection between the two intervals
    intersection = max(0, min(end1, end2) - max(start1, start2))
    # determine the union of the two intervals
    union = max(end1, end2) - min(start1, start2) - intersection
    # calculate the IoU as a percentage
    iou = (intersection / union) * 100
    return iou
```

```python
def overlap(start1, end1, start2, end2):
    # determine the length of each interval
    len1 = end1 - start1
    len2 = end2 - start2
    
    # determine the start and end points of the overlap
    overlap_start = max(start1, start2)
    overlap_end = min(end1, end2)
    
    # determine the length of the overlap
    overlap_len = max(0, overlap_end - overlap_start)
    
    # determine the percentage of overlap
    percent_overlap = (overlap_len / min(len1, len2)) * 100
    
    return percent_overlap
```

```python
def total_overlap_percent(start1, end1, ARR):
    total_overlap_len = 0
    total_len = 0
    
    for start2, end2 in ARR:
        # determine the length of each interval
        len1 = end1 - start1
        len2 = end2 - start2

        # determine the start and end points of the overlap
        overlap_start = max(start1, start2)
        overlap_end = min(end1, end2)

        # determine the length of the overlap
        overlap_len = max(0, overlap_end - overlap_start)

        # add the overlap length to the total overlap length
        total_overlap_len += overlap_len
        
        # add the length of the other interval to the total length
        total_len += len2
        
    # determine the total percentage overlap
    total_percent_overlap = (total_overlap_len / total_len) * 100
    
    # determine the total percentage by which the other intervals overlap our interval
    total_overlap_percent = 100 - total_percent_overlap
    
    return total_overlap_percent
```

```python
import numpy as np

def I2A(intervals,length=288):
    # Create an array of zeros with length 288
    arr = np.zeros(length)

    # Check if the input is a list of intervals or a single interval
    if isinstance(intervals, list):
        # Iterate over each interval
        for start, end in intervals:
            # Convert the start and end times to indices in the array
            start_idx = int(start)
            end_idx = int(end)

            # Set the values in the array to 1 for the points inside the interval
            arr[start_idx:end_idx] = 1
    elif isinstance(intervals, tuple):
        # Convert the start and end times to indices in the array
        start_idx = int(intervals[0])
        end_idx = int(intervals[1])

        # Set the values in the array to 1 for the points inside the interval
        arr[start_idx:end_idx] = 1
    else:
        raise ValueError("Intervals must be a list or a tuple")

    return arr
```

```python
import numpy as np

def f1_score(arr1, arr2):
    # Calculate true positives, false positives, and false negatives
    tp = np.sum(np.logical_and(arr1 == 1, arr2 == 1))
    fp = np.sum(np.logical_and(arr1 == 0, arr2 == 1))
    fn = np.sum(np.logical_and(arr1 == 1, arr2 == 0))

    # Calculate precision and recall
    precision = tp / (tp + fp+0.00001)
    recall = tp / (tp + fn+0.00001)

    # Calculate F1 score
    f1 = 2 * precision * recall / (precision + recall+0.0001)

    return f1
```

```python
def iou(box1, box2):
    """
    Calculates the intersection over union (IoU) between two boxes.
    
    Args:
        box1: A tuple of two coordinates representing the top-left and bottom-right corners of the first box.
        box2: A tuple of two coordinates representing the top-left and bottom-right corners of the second box.
    
    Returns:
        The IoU between the two boxes.
    """
    x1 = max(box1[0][0], box2[0][0])
    y1 = max(box1[0][1], box2[0][1])
    x2 = min(box1[1][0], box2[1][0])
    y2 = min(box1[1][1], box2[1][1])
    intersection = max(x2 - x1, 0) * max(y2 - y1, 0)
    area1 = (box1[1][0] - box1[0][0]) * (box1[1][1] - box1[0][1])
    area2 = (box2[1][0] - box2[0][0]) * (box2[1][1] - box2[0][1])
    union = area1 + area2 - intersection
    iou = intersection / union if union > 0 else 0
    return iou
        
        
for i in range(len(res)):
    if res.iloc[i].intervals!=None:
#         print(np.array(res.iloc[i].intervals))
#         if len(np.array(res.iloc[i].intervals)):
#         if not isinstance(seg, np.ndarray):
#         print((res.iloc[i].intervals, (res.iloc[i].startM, res.iloc[i].endM)))
#         print(res.iloc[i].intervals)
#         for I in list(res.iloc[i].intervals):
        
        print(iou(list(res.iloc[i].intervals), (res.iloc[i].startM, res.iloc[i].endM)))
#             print(I)
#         print(MAXIOU(res.iloc[i].intervals, (res.iloc[i].startM, res.iloc[i].endM)))
```

```python
def MAXIOU(intervals, e):
    """
    Calculate the Maximum Intersection over Union (MaxIoU) between a set of intervals and another interval.

    Args:
        intervals (list): A list of tuples representing the set of intervals (start time, end time).
        e (tuple): A tuple representing the input interval (start time, end time).

    Returns:
        float: The maximum IoU between the input interval and any of the intervals in the set.
    """
    max_iou = 0.0
    for interval in intervals:
        # Find the start and end times of the overlapping interval
        start_time = max(interval[0], e[0])
        end_time = min(interval[1], e[1])

        # Calculate the length of the overlapping interval
        overlap = max(0, end_time - start_time)

        # Calculate the length of the union of the two intervals
        union = (interval[1] - interval[0]) + (e[1] - e[0]) - overlap

        # Calculate the IoU
        iou = overlap / union

        # Update the maximum IoU value
        max_iou = max(max_iou, iou)
        
def TOTAL_IOU(intervals, interval):
    """
    Calculate the total IoU summarized from multiple overlaps between a set of intervals and another interval.

    Args:
        intervals (list): A list of tuples representing the set of intervals (start time, end time).
        interval (tuple): A tuple representing the input interval (start time, end time).

    Returns:
        float: The total IoU between the input interval and all intervals in the set.
    """
    total_iou = 0.0
    for i in intervals:
        iou = MAXIOU(intervals, i)
        total_iou += iou

    return total_iou

def calculate_iou_mean(df):
    ARRIOU = [TOTALIOU((row.startM, row.endM), (row.startW, row.endW)) for _, row in df.iterrows() if not math.isnan(TOTALIOU((row.startM, row.endM), (row.startW, row.endW)))]
    result = np.mean(np.array(ARRIOU) * 100)
    return ARRIOU, result
```
